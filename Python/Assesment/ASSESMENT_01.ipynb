#**8. How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?**

```python
state_bank_texas = banks[banks['Acquiring Institution'] == 'State Bank of Texas']
total_acquisitions = len(state_bank_texas)
texas_acquisitions = len(state_bank_texas[state_bank_texas['ST'] == 'TX'])
print("State Bank of Texas Acquisitions:")
print(f"Total Acquisitions: {total_acquisitions}")
print(f"Acquisitions in Texas: {texas_acquisitions}")
```

#**9. What is the most common city in California for a bank to fail in?**

```python
ca_failures = banks[banks['ST'] == 'CA']
most_common_ca_city = ca_failures['City'].mode().values[0]
print("Most Common City in California for Bank Failures:")
print(most_common_ca_city)
```
